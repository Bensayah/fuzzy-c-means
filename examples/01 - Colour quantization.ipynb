# Color quantization

```python
import numpy as np
from PIL import Image
from fcmeans import FCM
```

## Getting and rescaling the image

```python
original_image = Image.open('images/oarsman.jpeg')     # read image (oarsman at https://omadson.github.io/photos/)
copy_image = original_image.copy()                     # copy image
copy_image.thumbnail((200, 200), Image.ANTIALIAS)      # rescale image
N, M = copy_image.size                                 # get the number of columns (N) and rows (M)
copy_image                                             # show resized image
```

## Transforming image into a data set

```python
X = (
    np.asarray(copy_image)                             # convert a PIL image to np array
    .reshape((N*M, 3))                                 # reshape the image to convert each pixel to an instance of a data set
)
```

## Creating and fitting the model

```python
fcm = FCM(n_clusters=10)                               # create a FCM instance with 10 clusters
fcm.fit(X)                                             # fit the model
```

## Pixel quantization

```python
N, M = original_image.size                             # geting original image size

full_X = (
    np.asarray(original_image)                         # convert a PIL image to np array
    .reshape((N*M, 3))                                 # reshape the image to convert each pixel to an instance of a data set
)

labeld_X = fcm.predict(full_X)                         # get the label of each data point
transformed_X = fcm.centers[labeld_X]                  # pixel quantization into the centers
```

# Converting and saving image

```python
quatized_array = (
    transformed_X
    .astype('uint8')                                   # convert data points into 8-bit unsigned integers
    .reshape((M, N, 3))                                # reshape image
)

quatized_image = Image.fromarray(quatized_array)       # convert array into a PIL image object
quatized_image.save('images/oarsman_10colors.jpg')     # save image
```

## Final result
### Original / Quantized

```python
side_by_side = Image.fromarray(
    np.hstack([
        np.array(original_image),
        np.array(quatized_image)
    ])
)
side_by_side
```
